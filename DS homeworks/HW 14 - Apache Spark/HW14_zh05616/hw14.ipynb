```python
from pyspark import SparkConf, SparkContext
from pyspark.sql import SQLContext, SparkSession

#sc is the Spark Context already available at databricks, so we do not need to create it again.
sc = SparkContext()
sqlContext = SQLContext(sc)

# from pyspark.sql import SparkSession
sqlContext = SparkSession.builder.appName('abc').getOrCreate()
```

# Creating RDD (Resilient Distributed Dataset) - fundamental data structure of Spark

```python
products = sc.textFile("SalesData-1.txt")
```

```python
products.first()
```

```python
print("The data has {} columns".format(len(products.first().split("\t"))))
header = products.first().split("\t")
header
```

```python
head = products.first()
content = products.filter(lambda line: line != head)
```

```python
content.first()
```

```python
content.map(lambda line: len(line.split("\t"))).distinct().collect()[0]
```

# Example of RDD using Scala

```python
rdd1 = (content.filter(lambda line: line.split("\t"))
 .map(lambda line: (line.split("\t")))
)
rdd1.take(1)
```

# Converting RDD into Spark Dataframe

```python
df = rdd1.toDF(header)
df.show()
```

# Using Spark Dataframe syntax and SQL Table syntax

```python
# selecting two columns Name and Weight from original dataframe and convert into their proper datatype. Default is string
df1 = df.select(df.Name, df.Weight.cast("float"))
df1.show(5, truncate=False)
```

```python
# convert the above Dataframe into SQL Table
df1.createOrReplaceTempView("df_table")
sqlContext.sql(" SELECT * FROM df_table limit 5").show(truncate=False)
```

## 5. Visualizing Average List Price per Color and Average List Price per Size. Discuss/Explain your results.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pyspark.sql import functions as F
%matplotlib inline
```

```python
dfQ5a = df.select(df.Name,df.ProductNumber, df.Color,df.Size,df.ListPrice.cast("float"))
dfQ5b= dfQ5a.groupBy(df.Color).agg({"ListPrice":"average"}).collect()
color = []
averagePrice = []
for i in dfQ5b:
  if i["Color"]!= "NULL":
    color.append(i["Color"])
    averagePrice.append(i["avg(ListPrice)"])

plt.bar(color, averagePrice, width=0.5, color='blue')
plt.title('Average List Price per Color', fontsize=16)
plt.xlabel('Cp;pr', fontsize=12)
plt.ylabel('Average List Price', fontsize=12)
plt.show()
```

## Discussion
Most average list price for red color and least for white

```python
dfQ5c= dfQ5a.groupBy(df.Size).agg({"ListPrice":"average"}).collect()
size = []
averagePrice = []
for i in dfQ5c:
  if i["Size"]!= "NULL":
    size.append(i["Size"])
    averagePrice.append(i["avg(ListPrice)"])

plt.bar(size, averagePrice, width=0.5, color='blue')
plt.title('Average List Price per Size', fontsize=16)
plt.xlabel('Size', fontsize=12)
plt.ylabel('Average List Price', fontsize=12)
plt.xticks(size, rotation=45)
plt.show()
```

## Discussion
Most average list price for size 56 and least for size XL, M, L etc.

## 5. Visualizing Total Count per Size. Discuss/Explain your results.

```python
dfQ5d = df.select(df.Name,df.ProductNumber, df.Color,df.Size,df.ListPrice.cast("float"))
dfQ5e= dfQ5d.groupBy(df.Size).count().collect()
size = []
count = []
for i in dfQ5e:
  if i["Size"]!= "NULL":
    size.append(i["Size"])
    count.append(i["count"])

plt.bar(size, count, width=0.5, color='blue')
plt.title('Count per Size', fontsize=16)
plt.xlabel('Size', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(size, rotation=45)
plt.show()
```

## Discussion
 Most count for size 44 and least count for size 70

# Using Machine Learning Library (For Practice)

```python
dfall = df.select(df.Weight.cast("float"), df.Size.cast("float"), df.ListPrice.cast("float"))
dfall.show(5, truncate=False)
```

```python
dfall = dfall.na.drop("any")
```

```python
feature_columns = dfall.columns[:-1] # here we omit the final column for prediction
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=feature_columns,outputCol="features")
```

```python
data_2 = assembler.transform(dfall)
```

```python
data_2.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
algo = LinearRegression(featuresCol="features", labelCol="ListPrice")
```

```python
model = algo.fit(data_2)
```

```python
evaluation_summary = model.evaluate(data_2)
```

```python
evaluation_summary.rootMeanSquaredError
```
