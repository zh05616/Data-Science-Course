```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

#importing all the libraries that we need to compile the code

data=pd.read_csv("EmployeeAttrition.csv")

#reading the csv
```


We want to compare how much money men and women make each month. We'll do this by taking samples from both groups and doing some tests.

The null hypothesis is that men and women make about the same amount of money each month.

Our alternative hypothesis is that men might make more or less than women.

```python
male = data[(data['Gender']=='Male')]['MonthlyIncome']
female = data[(data['Gender']=='Female')]['MonthlyIncome']
stats.ttest_ind(male, female)

#this formula gives us the p value and other stats 
```

In summary, based on the obtained p-value of 0.22, which is greater than 0.05, we do not have sufficient evidence to reject the Null Hypothesis. 
This suggests that there is no statistically significant difference in the monthly income between males and females.

Null Hypothesis: The work life balance of males is less than the work life balance of females


Alternative Hypothesis: The work life balance of males is not less than the work life balance of females


```python
male = data[(data['Gender']=='Male')]['WorkLifeBalance']
female = data[(data['Gender']=='Female')]['WorkLifeBalance']
stats.ttest_ind(male, female)
```

In summary, as our p-value is 0.91 and therefore greater than 0.05, we do not have enough evidence to reject the Null Hypothesis. This implies that there is no statistically significant difference in work-life balance between males and females.

q3

Null Hypothesis: The years at company for single people is less than years at company for married people

Alternative Hypothesis: The years at company for single people is not less than years at company for married people

```python
single = data[(data['MaritalStatus']=='Single')]['YearsAtCompany']
married = data[(data['MaritalStatus']=='Married')]['YearsAtCompany']
stats.ttest_ind(single, married)
```


In summary, based on the obtained p-value of 0.01, which is less than 0.05, we have enough evidence to reject the Null Hypothesis and, as a result, accept the Alternative Hypothesis. This indicates that the years at the company for single people are not less than the years at the company for married people.

q6

Null Hypothesis: Gender and department are two variables that are not associated with each other


Alternative: Gender and department are associated with each other

```python
chisqt = pd.crosstab(data['Department'],data['Gender'], margins=True)
```

```python
value = np.array([chisqt.iloc[0][0:].values,
                  chisqt.iloc[1][0:].values])
```

```python
stats.chi2_contingency(value)
```

As evident from the provided values, the second value represents the p-value, which is 0.47. This result suggests that we should not reject the null hypothesis. Therefore, we can conclude that there is no significant association between Gender and Department; they are independent variables.
