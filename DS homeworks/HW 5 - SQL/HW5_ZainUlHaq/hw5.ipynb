```python
df1 = psql.read_sql('select * from employeeattrition1;', engine)
df2 = psql.read_sql('select * from employeeattrition1 e join employeeattrition2 d on e.EmployeeNumber = d.EmployeeNumber;', engine)
#selecting all rows and columns on both our dataframes
df2['jobrole'].value_counts
```

```python
df2['jobrole'].value_counts().plot(kind = 'bar', title = 'Workers in each role')
```

# Analyzing the data extracted from the database, we observe the presence of 9 distinct job roles making a total of 1470 entries. Within these 1470 entries, the graph highlights that approximately 22% of individuals are employed in the roles of sales executives, while research scientists and laboratory technicians closely follow. It's also worth noting that only 3% of the total workforce is engaged in Human Resources.

# Another observation is that the combined percentage of individuals working as managers, sales representatives, HR agents, and research directors is less compared to the percentage of individuals holding positions as sales executives.

```python
sns.boxplot(data = df2, x = 'gender', y = 'monthlyincome', palette = ['black', 'silver'])
plt.show()
```

# The box plot shows the distribution of monthly income among males and females. A close examination of the chart reveals that both genders receive nearly identical pay, with the median monthly income for females slightly more than that of males, at approximately $5,000. Additionally, the highest earners are female, with a monthly income of approximately $20,000, although this figure is very close with the highest male earnings.

# The presence of several outliers in the graph indicates that certain job roles, for both males and females, have significantly higher monthly incomes when compared to other positions. If we exclude these outliers, we find that the highest monthly salary for females is approximately $17,500, whereas for males, it is around $16,000. It hence becomes evident that both genders earn nearly equally, but specific job roles offer quite higher incomes than others.

```python
df2['department'].value_counts().plot(kind = 'pie', title = 'Department Distribution')
```

# The pie chart shows a breakdown of various departments within the dataset, revealing the existence of three distinct departments. The majority of employees are found in the Research and Development department, followed by the Sales department, and lastly, the Human Resources department. This observation suggests that the Research and Development department has the highest number of job opportunities, while the Human Resources department has the fewest.

# Furthermore, this chart implies that the Research and Development department has a greater variety of job roles compared to the other two departments. With Research and Development accounting for over 50% of the chart, it is seen that a huge chunk of the company's workforce is in this department.
