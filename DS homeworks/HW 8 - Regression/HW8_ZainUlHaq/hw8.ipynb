```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('EmployeeSalaryRegression.csv')
#reading the csv file
```

```python
df
```

# Q1

```python
X = sm.add_constant(df['TotalWorkingYears'])
Y = df['MonthlyIncome']
model1 = sm.OLS(Y,X).fit()
```

```python
model1.summary()
```




## For every working year of the employee, the average monthly income increases by 467.6584

```python
plt.plot(df['TotalWorkingYears'].values, df['MonthlyIncome'].values, 'ro')
plt.plot(df['TotalWorkingYears'].values, model1.fittedvalues.values, 'b')
plt.legend(['Data', 'Fitted model'])
plt.xlabel('TotalWorkigYears')
plt.ylabel('MonthlyIncome')
```
